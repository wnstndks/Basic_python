```python
import pandas as pd
import numpy as np

df=pd.read_csv('https://raw.githubusercontent.com/pykwon/python/master/testdata_utf8/titanic_data.csv',encoding='utf-8')
print(df.head(3))
print()
print(df.tail(3))
print()
print(df.columns)
```

```python
df['나이대'] = pd.cut(df['Age'], bins=[1, 20, 35, 60, 150], labels=["소년", "청년", "장년", "노년"])
df
```

```python
df.pivot_table(index='Sex',columns='Pclass',values='Survived')
```

```python
df.pivot_table(index=['Sex','나이대'],fill_value=0,columns='Pclass',values='Survived')
```
