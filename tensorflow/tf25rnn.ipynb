# RNN
RNN(Recurrent Neural Network)은 시계열 또는 순차 데이터를 예측하는 딥러닝을 위한 신경망 아키텍처입니다. RNN은 다양한 길이의 순차 데이터로 작업하고 자연 신호 분류, 언어 처리, 비디오 분석 등의 문제를 해결하는 데 특히 효과적입니다.


```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import SimpleRNN, LSTM, GRU, Dense
```

```python
model= Sequential() #순차적
# model.add(SimpleRNN(units=3, input_shape=(2,10))) #  가장 단순한 형태의 RNN이라고 하며, Vanilla RNN이라고 부르낟.
model.add(LSTM(units=3, input_shape=(2,10))) # SimpleRNN의 Long-Term 문제를 해결
# model.add(GRU(units=3, input_shape=(2,10))) # LSTM보다 단순한 구조를 가지나 성능은 우수함
print(model.summary())
# 하나의 셀은 입력이 들어가고 빠져나간다, 하나만 가지고 자연어 처리를 한다.
# 셀 하나만 가지고는 안되기에 셀을 옆에 자기 걸로 들어가고 이런게 또 있을수 있다.


print()
model=Sequential()
model.add(LSTM(units=3,batch_input_shape=(8,2,10), return_sequences=False)) # Many-To-one
print(model.summary())

print()
model=Sequential()
model.add(LSTM(units=3,batch_input_shape=(8,2,10), return_sequences=True)) # Many-To-Many
print(model.summary())

#학습을 시킬때 하나만 하면 비효율적이므로 8개씩 묶어서 작업을 하면 됨
```

4개의 숫자가 주어지면, 그 다음 숫자를 예측하기


```python
import numpy as np

x=[]
y=[]

for i in range(6):
    lst= list(range(i,i+4))
    # print(lst)
    x.append(list(map(lambda c:[c/10], lst)))
    y.append((i+4)/10)

x=np.array(x)
y=np.array(y)
# print(x)
# print(y)

model=Sequential([
    # SimpleRNN(units=10, activation='tanh', input_shape=[4,1]),
    LSTM(units=10, activation='tanh', input_shape=[4,1]),
    Dense(units=1) #값을 그 값 그대로 내보낸다.
])

model.compile(optimizer='adam', loss='mse')
model.summary()
model.fit(x=x,y=y,epochs=100, verbose=0)
print('예측값 : ', model.predict(x).flatten())
print('실제값 : ', y)
print()
print(model.predict(np.array([[[-0.1],[0.8],[0.9]]])))

```
