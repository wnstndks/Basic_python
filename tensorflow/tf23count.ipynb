단어의 빈도수로 자연어 특징 추출하기

CountVectorizer : 문서집합에서 단어 토큰을 생성하고 각 단어의 수를 세어 Bow 인코딩 벡터를 만든다.

TfidVectorizer:  CountVectorizer와 유사하지만 TF-IDF 방식으로 단어의 가중치를 조정한 BOW 인코딩 벡터를 만든다,

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

# CountVectorizer : 단순한 특징으로 corpus에서 단위별 등장횟수를 카운팅하여 수치벡터(BOW)화 함 - 글자 단위 또는 단어 단위 일수 있음
# 단위는 임의적, 단어, 글자, 자소, 문서, 문장 ....

content=['How to format my hard disk', 'Hard disk format format problems']

count_vec = CountVectorizer(analyzer='word', min_df=1)
print(count_vec)

train=count_vec.fit_transform(raw_documents=content)
print(train)
print(count_vec.get_feature_names_out()) # BOW vector : 토큰의 개수만큼 생성
# content
# ['disk' 'format' 'hard' 'how' 'my' 'problems' 'to']
#   0       1       2       3     4     5       6
print(train.toarray()) #배열의 순서대로 BOW에서 Word token이 몇번 나왔는지를 보여
print(train.format)
```

```python
# TfidVectorizer : CountVectorizer의 단점을 해결하기 위한 기법
# 특정 문서 내에서 특정 단어의 빈도인 TF(Term Fequency)와 전체 문서 내에서 특정 단어의 빈도인 DF(Documnet Fequency)
# 역수를 활용하여 어떠한 단어가 얼마나 중요한지를 나타낸 통계적 수치
# 문서에서 자주 나오는 단어에 높은 가중치를 주되 모든 문서에서 자주 나타나는 단어에 대해서는 패널티를 주는 방식으로 값 부여

tfidf_vec = TfidfVectorizer(analyzer='word', min_df=1)
train_idf=tfidf_vec.fit_transform(raw_documents=content)
print(train_idf)
print(tfidf_vec.get_feature_names_out()) #건수가 아니라 건수에 대한 확률값을 출력
print(train_idf.toarray()) #빈도에 대한 확률을 보여줌 -> 하나의 문장에 대해서 자주나오는 것은 확률값이 높다 그래서 이 비율을 이용해 확률값을 계산하는 방법
```
